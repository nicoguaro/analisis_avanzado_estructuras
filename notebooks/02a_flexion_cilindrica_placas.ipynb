## Caso empotrado-momento flector

```python
C1, C2, C3, C4 = symbols("C1 C2 C3 C4")
M = symbols("M")
```

```python
w2 = C1 + C2*x + C3*x**2 + C4*x**3
w2
```

```python
eqs = [
    w2.subs(x, 0),
    w2.diff(x).subs(x, 0),
    -D*w2.diff(x, 2) - M,
    -D*w2.diff(x, 3)]
eqs
```

```python
constantes = solve(eqs, [C1, C2, C3, C4])
constantes
```

```python
w2.subs(constantes)
```
